# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(0, 2*math.pi, 50)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('Funciones seno y coseno')
plt.show()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('Funciones seno y coseno')
plt.grid()
```

```python
plt.plot(x,y1,'g')
plt.plot(x,y2,'r*')
plt.title('Funciones seno y coseno')
plt.grid()
```

```python
plt.plot(x,y1,'m*')
plt.plot(x,y2,'b+')
plt.title('Funciones seno y coseno')
plt.grid()
```

```python
plt.plot(x,y1,'g-', linewidth=4)
plt.plot(x,y2,'ro--')
plt.title('Funciones seno y coseno')
plt.grid()
```

```python
plt.plot(x,y1,'b', linewidth=5)
plt.plot(x,y2,'m', linewidth=2)
plt.title('Funciones seno y coseno')
plt.grid()
```

```python
plt.plot(x,y1, color='#A976D7', linewidth=3, label='seno(x)')
plt.plot(x,y2, color='#76D7C4', linewidth=3, label='cos(x)')
plt.title('Funciones seno y coseno')
plt.grid()
```

```python
plt.plot(x,y1, color='#A976D7', linewidth=3, label='seno(x)')
plt.plot(x,y2, color='#76D7C4', linewidth=3, label='cos(x)')
plt.legend()
plt.title('Funciones seno y coseno')
plt.grid()
```

```python
plt.plot(x,y1, color='#A976D7', linewidth=3, label='seno(x)')
plt.plot(x,y2, color='#76D7C4', linewidth=3, label='cos(x)')
plt.legend(loc ='upper left')
plt.title('Funciones seno y coseno')
plt.grid()
```

```python
plt.plot(x,y1, color='#A976D7', linewidth=3, label='seno(x)')
plt.plot(x,y2, color='#76D7C4', linewidth=3, label='cos(x)')
plt.legend(loc ='upper left')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funciones seno y coseno')
plt.grid()
```

```python
plt.plot(x,y1, color='#A976D7', linewidth=3, label='seno(x)')
plt.plot(x,y2, color='#76D7C4', linewidth=3, label='cos(x)')
plt.legend(loc ='best', fontsize=14)
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$f(x)$', fontsize=18)
plt.title('Funciones seno y coseno', fontsize=20)
plt.grid()
```

## Diagrama barras

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
materias = ['Matemticas', 'Lectura', 'Ingles', 'Biologia', 'Sociales']
nota = []
for materia in materias:
  x = float(input(f'Ingrese la nota en {materia}:'))
  nota.append(x)
```

```python
plt.bar(materias, nota)
```

```python
plt.bar(materias, nota)
plt.title('Grficas notas por materia', fontsize=16)
plt.ylabel('Nota')
```

```python
plt.bar(materias, nota)
plt.axhline(y=3, color='r')
plt.title('Grficas notas por materia', fontsize=16)
plt.ylabel('Nota')
```

```python
colores = ['#A976D7', '#76CBD7', '#80D776', '#D776B4', '#D7B776']
i = 0
for materia in materias:
  plt.bar(materia, nota[i], color=colores[i])
  i +=1
plt.axhline(y=3, color='r', lw=2)
plt.title('Grficas notas por materia', fontsize=16)
plt.ylabel('Nota', fontsize=16)
plt.xlabel('Materias', fontsize=16)
```

## Barras dobles

```python
materias = ['Matemticas', 'Lectura', 'Ingles', 'Biologia', 'Sociales']
nota = []
nota2 = []
for materia in materias:
  x = float(input(f'Ingrese la promedio del salon 7-2 en {materia}:'))
  nota.append(x)
  x = float(input(f'Ingrese la promedio del salon 7-3 en {materia}:'))
  nota2.append(x)
```

```python
indice_barras = np.arange(len(nota))
indice_barras
```

```python
indice_barras + 0.25
```

```python
width_bar = 0.25
 
plt.bar(indice_barras, nota, width=width_bar, label='Salon 7-2', color='#549EB5')
plt.bar(indice_barras + width_bar, nota2, width=width_bar, label='Salon 7-3', color='#B56B54')
plt.legend(loc='best')

plt.xticks(indice_barras + width_bar, materias)
 
plt.ylabel('Nota')
plt.title('Promedio de notas por curso')
 
plt.axhline(y=3.5, color='r', lw=2)
plt.show()
```

```python
x = 0.9 * np.random.rand(50)
y = 0.9 * np.random.rand(50)

plt.scatter(x, y, marker='*',s = 150, c = 'r')
plt.title('Datos')
plt.show()
```

```python
x = 0.9 * np.random.rand(50)
y = 0.9 * np.random.rand(50)
k = 0
for i in x:
  if i < 0.4:
    plt.scatter(x[k], y[k], marker='^',s = 140, c = 'r')
  else:
    plt.scatter(x[k], y[k], marker='o',s = 140, c = 'b')
  k+=1


plt.title('Datos')
plt.show()
```

```python
x = 0.9 * np.random.rand(50)
y = 0.9 * np.random.rand(50)
k = 0
for i in x:
  if i < 0.4:
    if y[k]<0.6:
      plt.scatter(x[k], y[k], marker='^',s = 140, c = '#18D53D')
    else:
      plt.scatter(x[k], y[k], marker='^',s = 80, c = '#18D5CF')
  else:
    if y[k]<0.6:
      plt.scatter(x[k], y[k], marker='o',s = 90, c = '#D518CC')
    else:
      plt.scatter(x[k], y[k], marker='*',s = 180, c = '#D51868')
  k+=1


plt.title('Datos')
plt.show()
```
